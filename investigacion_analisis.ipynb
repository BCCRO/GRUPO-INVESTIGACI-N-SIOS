```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords

# Cargar los datos desde un archivo CSV o DataFrame de pandas
data = pd.read_csv('/content/respuestas_investigacion.csv')

# Inicializar el analizador de sentimientos de VADER
sid = SentimentIntensityAnalyzer()

# Obtener lista de stopwords en español
stop_words = set(stopwords.words('spanish'))

# Crear un diccionario para almacenar los resultados
resultados = {}
sentimientos_totales = []
palabras_totales = Counter()

# Iterar sobre cada columna (pregunta)
for columna in data.columns:
    respuestas = data[columna].tolist()

    # Realizar análisis de sentimientos y contar la frecuencia de palabras clave
    sentimientos = []
    palabras_clave = Counter()

    for respuesta in respuestas:
        # Análisis de sentimientos
        scores = sid.polarity_scores(str(respuesta))
        sentimientos.append(scores['compound'])
        sentimientos_totales.append(scores['compound'])

        # Tokenización y eliminación de stopwords
        palabras = nltk.word_tokenize(str(respuesta).lower())
        palabras_filtradas = [palabra for palabra in palabras if palabra.isalnum() and palabra not in stop_words]
        palabras_clave.update(palabras_filtradas)
        palabras_totales.update(palabras_filtradas)

    # Guardar resultados en el diccionario
    resultados[columna] = {'sentimientos': sentimientos, 'palabras_clave': palabras_clave}

# Mostrar resultados y generar word clouds y gráficos de barras por pregunta
for columna, datos in resultados.items():
    print(f"Pregunta: {columna}")

    # Calcular la media de los sentimientos para la pregunta
    media_sentimientos = sum(datos['sentimientos']) / len(datos['sentimientos'])

    # Determinar la conclusión para la pregunta
    if media_sentimientos > 0:
        conclusion = "Las respuestas son mayoritariamente positivas."
    elif media_sentimientos < 0:
        conclusion = "Las respuestas son mayoritariamente negativas."
    else:
        conclusion = "Las respuestas son neutrales."

    print("Análisis de sentimientos:")
    for i, sentimiento in enumerate(datos['sentimientos']):
        print(f"Respuesta {i+1}: Sentimiento = {sentimiento}")

    print("\nFrecuencia de palabras más comunes:")
    for palabra, frecuencia in datos['palabras_clave'].most_common(10):
        print(f"{palabra}: {frecuencia}")

    # Generar word cloud de las palabras más frecuentes
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(datos['palabras_clave'])
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f"Word Cloud de las palabras más frecuentes - Pregunta: {columna}")
    plt.axis('off')
    plt.show()

    # Generar gráfico de barras de las palabras más frecuentes por pregunta
    palabras = [item[0] for item in datos['palabras_clave'].most_common(10)]
    frecuencias = [item[1] for item in datos['palabras_clave'].most_common(10)]

    plt.figure(figsize=(12, 6))
    plt.bar(palabras, frecuencias, color='#00BFBF')
    plt.title(f"Frecuencia de las palabras más comunes - Pregunta: {columna}")
    plt.xlabel('Palabras')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.show()

    print(f"\nConclusión para la pregunta '{columna}': {conclusion}")
    print(f"Media de los sentimientos: {media_sentimientos:.3f}\n")

# Calcular la media de los sentimientos totales
media_sentimientos = sum(sentimientos_totales) / len(sentimientos_totales)

# Calcular el número de sentimientos positivos, negativos y neutrales
positivos = sum(1 for sentimiento in sentimientos_totales if sentimiento > 0)
negativos = sum(1 for sentimiento in sentimientos_totales if sentimiento < 0)
neutrales = sum(1 for sentimiento in sentimientos_totales if sentimiento == 0)

# Conclusión general
if media_sentimientos > 0:
    conclusion_general = "En general, las respuestas son positivas."
else:
    conclusion_general = "En general, las respuestas son negativas."

print(f"\nConclusión general: {conclusion_general}")
print(f"Media de los sentimientos: {media_sentimientos:.3f}")

# Gráficos de torta y de barras para los sentimientos en general
sentimientos = ['Positivos', 'Negativos', 'Neutrales']
cantidad = [positivos, negativos, neutrales]

# Gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(cantidad, labels=sentimientos, autopct='%1.1f%%', startangle=140, colors=['#511F73', '#26A699', '#F2BE22'])
plt.title('Distribución de Sentimientos en General')
plt.show()

# Generar gráfico de barras para las palabras más frecuentes en general
palabras_comunes = palabras_totales.most_common(12)
palabras = [item[0] for item in palabras_comunes]
frecuencias = [item[1] for item in palabras_comunes]

plt.figure(figsize=(12, 6))
plt.bar(palabras, frecuencias, color='#00BFBF')
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```
